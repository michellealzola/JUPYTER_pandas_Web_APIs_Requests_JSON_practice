```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://data.calgary.ca/resource/35ra-9556.json'
```

```python
response = requests.get(url)
```

```python
data = response.json()
```

```python
data
```

```python
df_data = pd.DataFrame(data)
```

```python
df_data
```

```python
# Filter Multi-vehicle incident
```

```python
df_multi_vehicle_incident = df_data[df_data['description'] == 'Multi-vehicle incident.']
```

```python
df_multi_vehicle_incident
```

```python
# Count all traffic incidents in NE, NW, SE and SW
quadrants = ['NE', 'NW', 'SE', 'SW']
df_quadrant_count = pd.DataFrame(quadrants)
df_quadrant_count.rename(columns={0: 'Quadrants'}, inplace=True)
df_quadrant_count

```

```python
counts = []
for quadrant in quadrants:
    count = df_data[df_data['quadrant'] == quadrant].agg('quadrant').count()
    counts.append(count)

df_quadrant_count['Traffic Incident Counts'] = counts
```

```python
counts
```

```python
df_quadrant_count.set_index('Quadrants', inplace=True)
```

```python
df_quadrant_count
```

```python
df_quadrant_count.plot(kind='barh')
plt.show()
```

```python
# Show number of incidents per location

df_data.incident_info.count()
```

```python
from collections import Counter
```

```python
count = pd.Series(df_data['incident_info'].str.split('\n').map(Counter).sum())
```

```python
df_count = pd.DataFrame(count)
df_count.rename(columns={0 : 'Number_of_incidents'}, inplace=True)
```

```python
count_more_than_3 = df_count[df_count.Number_of_incidents >= 3]
```

```python
count_more_than_3.plot(kind='barh')
plt.ylabel('Location')
plt.show()
```
